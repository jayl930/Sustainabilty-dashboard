```python
import pandas as pd
```

```python
nodes_df = pd.read_csv('neo4j_nodes.csv')
relationships_df = pd.read_csv('neo4j_relationships.csv')
researchers_relationships_df = pd.read_csv('neo4j_researchers_relationships.csv')
people_articles_relationships_df = pd.read_csv('neo4j_people_articles_relationships.csv')
research_papers_df = pd.read_csv('research_papers_cleaned.csv')

articles_authors_df = pd.merge(
    people_articles_relationships_df,
    researchers_relationships_df,
    how='left',
    left_on='From',  # faculty name in people_articles_relationships
    right_on='From'  # faculty name in researchers_relationships
)

# Step 3: Merge with nodes_df to get organization details
articles_authors_org_df = pd.merge(
    articles_authors_df,
    nodes_df,
    how='left',
    left_on='To_y',  # organization name in researchers_relationships
    right_on='Name'  # organization name in nodes
)

# Step 4: Merge with research_papers_cleaned to add author and organization information
final_df = pd.merge(
    research_papers_df,
    articles_authors_org_df,
    how='left',
    left_on='uuid',  # article UUID in research_papers_cleaned
    right_on='To_x'  # article UUID in people_articles_relationships
)

# Step 5: Save the final merged DataFrame into a new CSV file
final_df.to_csv('merged_research_papers.csv', index=False)
```

```python
filtered_organizations = ['Gies College of Business', 'Accountancy', 'Business Administration', 'Finance']
filtered_df = df[df['organization'].isin(filtered_organizations)]

# Step 2: Create separate DataFrames for each organization and remove duplicate articles
gies_df = filtered_df[filtered_df['organization'] == 'Gies College of Business'].drop_duplicates(subset='uuid')
accountancy_df = filtered_df[filtered_df['organization'] == 'Accountancy'].drop_duplicates(subset='uuid')
business_admin_df = filtered_df[filtered_df['organization'] == 'Business Administration'].drop_duplicates(subset='uuid')
finance_df = filtered_df[filtered_df['organization'] == 'Finance'].drop_duplicates(subset='uuid')

# Extract UUIDs as sets for comparison
gies_uuids = set(gies_df['uuid'])
accountancy_uuids = set(accountancy_df['uuid'])
business_admin_uuids = set(business_admin_df['uuid'])
finance_uuids = set(finance_df['uuid'])

# Step 3: Find intersections (common articles)
common_gies_accountancy = gies_uuids.intersection(accountancy_uuids)
common_gies_business_admin = gies_uuids.intersection(business_admin_uuids)
common_gies_finance = gies_uuids.intersection(finance_uuids)
common_accountancy_business_admin = accountancy_uuids.intersection(business_admin_uuids)
common_accountancy_finance = accountancy_uuids.intersection(finance_uuids)
common_business_admin_finance = business_admin_uuids.intersection(finance_uuids)

# Find common articles across all four organizations
common_all_four = gies_uuids.intersection(accountancy_uuids, business_admin_uuids, finance_uuids)

# Step 4: Output the results
print("Common Articles between Gies and Accountancy:", len(common_gies_accountancy))
print("Common Articles between Gies and Business Administration:", len(common_gies_business_admin))
print("Common Articles between Gies and Finance:", len(common_gies_finance))
print("Common Articles between Accountancy and Business Administration:", len(common_accountancy_business_admin))
print("Common Articles between Accountancy and Finance:", len(common_accountancy_finance))
print("Common Articles between Business Administration and Finance:", len(common_business_admin_finance))
print("Common Articles among all four organizations:", len(common_all_four))
```

```python
common_with_others = gies_uuids.intersection(accountancy_uuids.union(business_admin_uuids, finance_uuids))

# Step 4: Remove common articles from Gies DataFrame
gies_df_unique = gies_df[~gies_df['uuid'].isin(common_with_others)]

# Output the number of articles in the updated Gies DataFrame
print("Number of unique articles in Gies after removing overlaps:", gies_df_unique.shape[0])
```

```python
import matplotlib.pyplot as plt

# Function to create bar chart for label distribution for a given DataFrame and organization name
def plot_label_distribution(df, org_name):
    # Step 1: Group by 'label' and count the number of papers for each label
    label_counts = df['label'].value_counts()
    
    # Step 2: Create a bar chart for the count of papers by label
    plt.figure(figsize=(10, 6))
    label_counts.plot(kind='bar', color='skyblue')

    # Customize the plot
    plt.title(f'Number of Research Papers by Label for {org_name}')
    plt.xlabel('Research Label')
    plt.ylabel('Number of Papers')
    plt.xticks(rotation=45)
    plt.tight_layout()

    # Show the plot
    plt.show()

# Plot for Gies College of Business
plot_label_distribution(gies_df_unique, 'Gies College of Business')

# Plot for Accountancy
plot_label_distribution(accountancy_df, 'Accountancy')

# Plot for Business Administration
plot_label_distribution(business_admin_df, 'Business Administration')

# Plot for Finance
plot_label_distribution(finance_df, 'Finance')
```
